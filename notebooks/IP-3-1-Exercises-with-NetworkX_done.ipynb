# Exercises with NetworkX

1. Graph with x nodes at random position
2. Graph with x nodes at random position and all nodes are connected with edges
3. Graph with x nodes at random position, nodes with given distance are connected


Following the results from the lecture discussion


```python
import math
import networkx as nx
import random
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# 1. Graph with x nodes at random position

num_nodes: int = 10

G = nx.Graph()
G.add_nodes_from(range(num_nodes))

nx.set_node_attributes(G, {node: (random.random(), random.random()) for node in G.nodes()}, 'pos')


nx.draw(G, nx.get_node_attributes(G, 'pos'), with_labels=True, node_size=500, node_color="skyblue")
plt.show()
```

```python
#2. Graph with x nodes at random position and all nodes are connected with edges

num_nodes: int = 10

G = nx.complete_graph(num_nodes)

nx.set_node_attributes(G, {node: (random.random(), random.random()) for node in G.nodes()}, 'pos')

nx.draw(G, nx.get_node_attributes(G, 'pos'), with_labels=True, node_size=500, node_color="skyblue")
plt.show()
```

```python
# 3. Graph with x nodes at random position, nodes with given distance are connected

num_nodes: int = 10
max_connected_distance: float = 0.3

G = nx.Graph()
G.add_nodes_from(range(num_nodes))

nx.set_node_attributes(G, {node: (random.random(), random.random()) for node in G.nodes()}, 'pos')

positions = nx.get_node_attributes(G, 'pos')
edges = [(n1, n2) for n1, p1 in positions.items() 
         for n2, p2 in positions.items() 
         if n1 < n2 and math.dist(p1, p2) < max_connected_distance]
G.add_edges_from(edges)

nx.draw(G, nx.get_node_attributes(G, 'pos'), with_labels=True, node_size=500, node_color="skyblue")
plt.show()
```
